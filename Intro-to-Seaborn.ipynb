```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

## 1. Plot a scatterpot  b/w total_bill and tip

```python
sns.relplot(x='total_bill', y='tip', kind='scatter', data=data)
```

### The hue parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=data)
```

### The style parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex',
            size='size', data=data)
```

## 2. Plotting with Categorical Data

### A. Categorical Scatterplot

#### a. Stripplot
#### b. Swarmplot

```python
# Categorical Scatterplot prerequists
# x-axis --> Categorical data
# y-axis --> Numerical data
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
# a. Stripplot
sns.catplot(x='day', y='tip', kind='strip', jitter=0, data=tips)
```

```python
# b. Swarmplot
sns.catplot(x ='day', y='tip', kind='swarm', data=tips)
```

```python
sns.catplot(x='day', y='tip', kind='swarm', hue='sex', data=tips)
```

### B. Categorical distribution plots

#### a. Boxplot
#### b. Violinplot

```python
# a. boxplot
sns.boxplot(tips['tip'])
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='box', data=tips)
```

```python
# b. violenplots
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

### C. Categorical estimate plots

#### a. Pointplot
#### b. Barplot
#### c. Countplot

```python
# Barplot
sns.catplot(x='smoker', y='total_bill', hue='sex', kind='bar', data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex',
            estimator=np.median, kind='bar', data=tips)
```

```python
# countplot
sns.catplot(x='smoker', y='total_bill', hue='sex', kind='bar', data=tips)
```

### D. Categorical Categorical plots

#### a. Heatmap
#### b. Clustermaps

```python
# heatmap
flights = sns.load_dataset('flights')
flights.head()
```

```python
x=flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x)
```

```python
# value driven heatmap
sns.heatmap(x, linewidths=0.3, annot=True, fmt='d')
```

```python
# b. Clustermaps
sns.clustermap(x)
```

```python
# joint plot
tips=sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# first type of jointplot i.e hexplot
sns.jointplot(x='total_bill', y='tip', kind='hex', data=tips)
```

```python
# second type of jointplot i.e kdeplot
sns.jointplot(x='total_bill', y='tip', kind='kde', data=tips)
```

```python
# third type of jointplot i.e regressionplot
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
```

## 3. Distplot

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

```python
# plotting the distplot
sns.distplot(titanic['age'])
```

```python
# plotting only the histogram
sns.distplot(titanic['age'], kde=False)
```

```python
# plotting only kde
sns.distplot(titanic['age'], hist=False)
```

```python
# plotting only the rugplot
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
```
