```python
import numpy as numpy 
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Basic Graph

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y) # plotting the values
plt.xlabel("X - axis") # for naming the X axis
plt.ylabel("Y - axis") # naming the y axis
plt.title("Our First Plot!") #  namng the title
plt.show()
```

## 2. Coustomize


### Changing the font

```python
plt.plot(x,y) 
 
plt.title("Our First Plot!", fontdict = {'fontname': 'Comic Sans MS'})

plt.xlabel("X - axis") 
plt.ylabel("Y - axis")

plt.xticks([1,2,3,4,5]) # for removing the decimal values and remaining 
plt.yticks([1,2,3,4,5]) # the integer values.

plt.show()
```

### Adding the legend, color and more to the data

```python
import numpy as np

plt.plot(x,y, label='2x', color = 'red', marker='.',linestyle='--',
         markersize=10, markeredgecolor='blue')
plt.title("Our First Plot!", fontdict = {'fontname': 'Comic Sans MS'})

plt.xlabel("X - axis") 
plt.ylabel("Y - axis")

plt.xticks([1,2,3,4,5]) 
plt.yticks([1,2,3,4,5])

plt.legend() # legend is used the add identity in the data.

plt.show()
```

### Using numpy to complicate data, because real world data is not that easy

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]


# resize your graph
plt.figure(figsize=(5,3), dpi=300)


plt.plot(x,y, label='2x', color = 'red', marker='.',linestyle='--',
         markersize=10, markeredgecolor='blue')

# Numpy
x2 = np.arange(0,4.5, 0.5)
plt.plot(x2, x2**2, label='x^2')

plt.title("Our First Plot!", fontdict = {'fontname': 'Comic Sans MS'})
plt.xlabel("X - axis") 
plt.ylabel("Y - axis")

plt.xticks([0,1,2,3,4]) 
plt.yticks([0,2,4,6,8,])

plt.legend() 

plt.show()
```

## 3. Plotting bar graphs

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

plt.yticks([0,1,2,3,4])

# styling the bars
bars[0].set_hatch('/')
bars[1].set_hatch('-')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```

## 4. Real World Examples

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
# plotting the graphs
plt.plot(gas.Year, gas.USA)
plt.plot(gas.Year, gas.Canada)
plt.xlabel('Year')
plt.ylabel('Prices')
plt.show()
```

```python
# plotting and styling multiple graphs in one plot

plt.title('Country Wise Gas Prices')

plt.xlabel('Year')
plt.ylabel('Prices')

plt.plot(gas.Year, gas.USA,  label='US',
         color = 'green', marker='.', linestyle='--')
plt.plot(gas.Year, gas.Canada, label='Canada',
         color = 'blue', marker='.', linestyle='--')
plt.plot(gas.Year, gas['South Korea'], label='South Korea',
         color = 'red', marker='.', linestyle='--')
plt.plot(gas.Year, gas.Australia, label='Australia',
         color = 'brown', marker='.', linestyle='--')


plt.xticks(gas.Year[::3])

plt.legend()

plt.show()

# for saving the graph directly to your machine just write :
# plt.savefig('Country Wise Gas Prices.png', dpi=300)
```

### Load Fifa Data

```python
fifa = pd.read_csv('PlayerPersonalData.csv')
fifa.head()
```

## 5. Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.xlabel('Number of Players')
plt.ylabel('Skill Level')
plt.title('Distribution of Players Skills n FIFA 2018')

plt.show()
```

## 6. Piechart

```python
# select the countries to be plotted
Argentina = fifa.loc[fifa['Nationality'] =='Argentina'].count()[0]
Germany = fifa.loc[fifa['Nationality'] =='Germany'].count()[0]
Brazil = fifa.loc[fifa['Nationality'] =='Brazil'].count()[0]
Uruguay = fifa.loc[fifa['Nationality'] =='Uruguay'].count()[0]

# labelling and styling the pie chart
labels = ['Argentina', 'Germany', 'Brazil', 'Uruguay']
colors = ['orange', 'indigo', 'green', 'red']

# plotting the plot
plt.pie([Argentina, Germany, Brazil, Uruguay], 
        labels=labels,
        colors=colors)

plt.title('Nationality (No. of players from Countries)')
plt.show()
```

## 7. Box Plot

```python
plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid CF']['Overall']
paris =  fifa.loc[fifa.Club == 'Paris Saint-Germain']['Overall']

plt.title('Team Comparison')
plt.ylabel('FIFA Overall Ratings')
labels = ['FC Barcelona', 'Real Madrid CF', 'Paris Saint-Germain']

boxes = plt.boxplot([barcelona, madrid, paris], labels=labels,
                   patch_artist=True, medianprops={'linewidth':2})

# style the box
for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2)
    
# fill the color inside the box
    box.set(facecolor='#e0e0e0')
    
plt.show()
```
